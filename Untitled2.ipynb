```python
import torch
import torchvision
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torchvision.transforms import functional as F
from torch.utils.data import DataLoader, Dataset

# Custom Dataset 정의
class CustomDataset(Dataset):
    def __init__(self, image_paths, annotations, transform=None):
        self.image_paths = image_paths
        self.annotations = annotations
        self.transform = transform

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        img = Image.open(img_path).convert("RGB")
        target = self.annotations[idx]

        if self.transform:
            img = self.transform(img)

        return img, target

    def __len__(self):
        return len(self.image_paths)

# 모델 초기화
model = fasterrcnn_resnet50_fpn(pretrained=True)

# 마지막 레이어 수정 (클래스 수 정의: 손가락, 손, 양손 + 작업 구분)
num_classes = 4  # 예: Background, 손가락, 손, 양손
in_features = model.roi_heads.box_predictor.cls_score.in_features
model.roi_heads.box_predictor = torchvision.models.detection.faster_rcnn.FastRCNNPredictor(in_features, num_classes)

device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
model.to(device)
```

```python
from PIL import Image
from torch.utils.data import Dataset, DataLoader

# Custom Dataset 정의
class CustomDataset(Dataset):
    def __init__(self, image_paths, annotations, transform=None):
        self.image_paths = image_paths  # 이미지 경로 리스트
        self.annotations = annotations  # 이미지에 대응하는 라벨 정보
        self.transform = transform  # 변환 함수

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        img = Image.open(img_path).convert("RGB")
        target = self.annotations[idx]

        if self.transform:
            img = self.transform(img)

        return img, target

    def __len__(self):
        return len(self.image_paths)

# 이미지 경로와 라벨 정보 정의
image_paths = ["dydwjq.jpg"]  # 예시 이미지 경로
annotations = [
    {
        "boxes": [[10, 20, 50, 60], [30, 40, 70, 80]],  # 첫 번째 이미지의 Bounding Box 좌표
        "labels": [1, 2]  # 클래스 라벨 (1=손가락, 2=손)
    },
    {
        "boxes": [[15, 25, 55, 65]],  # 두 번째 이미지의 Bounding Box
        "labels": [3]  # 클래스 라벨 (3=양손)
    }
]

# 데이터셋 및 데이터 로더 초기화
train_dataset = CustomDataset(image_paths, annotations, transform=None)
train_loader = DataLoader(train_dataset, batch_size=2, shuffle=True, collate_fn=lambda x: tuple(zip(*x)))

# 데이터 확인
for images, targets in train_loader:
    print("Images:", images)
    print("Targets:", targets)
```

```python
import matplotlib.pyplot as plt

# 평가 함수
def visualize_prediction(model, image_path, device):
    model.eval()
    img = Image.open(image_path).convert("RGB")
    img_tensor = F.to_tensor(img).unsqueeze(0).to(device)

    with torch.no_grad():
        predictions = model(img_tensor)

    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    for box, label in zip(predictions[0]['boxes'], predictions[0]['labels']):
        x1, y1, x2, y2 = box.cpu().numpy()
        plt.gca().add_patch(plt.Rectangle((x1, y1), x2-x1, y2-y1, edgecolor='red', fill=False, linewidth=2))
        plt.text(x1, y1, str(label), color='blue', fontsize=12)

    plt.show()

# 테스트
visualize_prediction(model, "dydwjq.jpg", device)
```
